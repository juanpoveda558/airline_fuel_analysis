```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Esto es para que los gráficos se vean bien dentro del notebook
%matplotlib inline 

print(f"Pandas versión {pd.__version__} cargado.")
print(f"Seaborn versión {sns.__version__} cargado.")

```

```python
# Cargar el archivo blanco que ya tienes en la carpeta
df = pd.read_csv("states_2022-06-27-00.csv", engine='pyarrow')
df.head()
```

```python
# Contamos cuántas filas hay por cada avión y vemos los 5 con más registros
print("Top 5 aeronaves con más datos:")
print(df['icao24'].value_counts().head(5))
```

```python
# Sustituye 'xxxxxx' por el código icao24 que más te haya gustado de la lista anterior
target_icao = '34718e' 

# Creamos el dataset ligero
df_single = df[df['icao24'] == target_icao].copy()

# Ordenamos por tiempo para que las gráficas tengan sentido cronológico
# Nota: Si tu columna de tiempo se llama 'time' o 'timestamp'
df_single = df_single.sort_values(by='time') 

print(f"Dataset creado para la aeronave {target_icao} con {len(df_single)} registros.")
```

```python
# Guardamos este subconjunto en un archivo nuevo
df_single.to_csv("single_aircraft_data.csv", index=False)
print("Archivo 'single_aircraft_data.csv' guardado con éxito.")
```
